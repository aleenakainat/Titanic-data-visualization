# **Titanic Data Visualization**

```python
# 1. Install & Import Required Libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 2. Load Titanic Dataset
# Seaborn has built-in Titanic dataset
df = sns.load_dataset('titanic')

# Quick look at the data
print("Data Shape:", df.shape)
print(df.head())
```

```python
# 3. Basic Data Cleaning (Optional)
# Drop rows with too many missing values for simplicity
df = df.dropna(subset=['age', 'fare', 'embarked', 'sex', 'class'])
```

```python
# 4. Quick Summary Stats (Extra Insight)
print("\nAverage Fare by Class:")
print(df.groupby('class')['fare'].mean())
print("\nSurvival Rate by Class:")
print(df.groupby('class')['survived'].mean())
```

```python
# 5. Bar Chart: Passenger Count by Class
plt.figure(figsize=(7,5))
sns.countplot(x='class', data=df, palette='pastel')
plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# 6. Scatter Plot: Age vs Fare colored by Survival
plt.figure(figsize=(7,5))
sns.scatterplot(x='age', y='fare', hue='survived', data=df, palette='coolwarm', alpha=0.7)
plt.title('Age vs Fare (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# 7. Boxplot: Fare Distribution by Embarked Port
plt.figure(figsize=(7,5))
sns.boxplot(x='embarked', y='fare', data=df, palette='Set2')
plt.title('Fare Distribution by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Fare')
plt.tight_layout()
plt.show()
```

```python
# 8. Heatmap: Correlation Between Numeric Features
# Select numeric columns only
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# 9. Extra Plot: Survival Count by Gender (Optional)
plt.figure(figsize=(6,4))
sns.countplot(x='sex', hue='survived', data=df, palette='pastel')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.tight_layout()
plt.savefig('survival_by_gender.png', dpi=300)
plt.show()
```
